```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
event_particles = pd.read_csv("../data/particles-per-event.txt", sep=" ", header=None)
```

```python
event_particles
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.hist(event_particles, bins=np.arange(0, 900, 100))
ax.set_xlabel("Generated Particles in Event")
ax.set_ylabel("Events")
ax.set_title("Input Generated Events (Pythia 8, CoM=13TeV, Jet $p_t>20GeV$")
plt.savefig("../data/event-content.png", dpi=150, bbox_inches='tight')
```
